### Iterators
Iterators are advanced Python concepts that allow for efficient looping and memory management. Iterators provide a way to access elements of a collection sequentially without exposing the underlying structure.


```python
my_list=[1,2,3,4,5,6]
for i in my_list:
    print(i)
```

```python
type(my_list)
```

```python
print(my_list)
```

```python
## Iterator
iterator = iter(my_list)
##it usese lazy evaluation, it will not create a new list in memory, 
# it will create an iterator object that will generate the values on demand.                               
print(type(iterator))
```

```python
iterator
```

```python
## Iterate through all the element


## first element will be displayed
next(iterator)
```

```python
iterator=iter(my_list)
## on calling next() on an empty iterator, it will raise StopIteration exception
next(iterator)
```

```python


try:
    print(next(iterator))
    ## this will raise StopIteration exception
    ## after the last element, it will raise StopIteration exception
except StopIteration:
    print("There are no elements in the iterator")
```

```python
# String iterator
my_string = "Hello"
string_iterator = iter(my_string)
## same goes for string, it will create an iterator object that will generate the characters on demand.
print(next(string_iterator))  # Output: H
print(next(string_iterator))  # Output: e
```
