#### Abstraction
Abstraction is the concept of hiding the complex implementation details and showing only the necessary features of an object. This helps in reducing programming complexity and effort.

```python
from abc import ABC,abstractmethod

## Abstract base cclass
class Vehicle(ABC): ## abract class always inherit from ABC
    def drive(self):
        print("The vehicle is used for driving")

    @abstractmethod 
    ##  the abrast method is a method that is declared but contains no implementation. 
    # It is meant to be overridden by subclasses. 
    ## Car class must implement the start_engine method otherwise it will raise an error.
    def start_engine(self):
        pass



class Car(Vehicle):
   def start_engine(self):
        print("The car engine is starting")


def operate_vehicle(vehicle):
    vehicle.start_engine()
    vehicle.drive()

  
car = Car()
operate_vehicle(car)





```
